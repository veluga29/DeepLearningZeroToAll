```python
import tensorflow as tf
import numpy as np
```

```python
sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")
```

```python
char_set = list(set(sentence))  # index -> char
char_dic = {w: i for i, w in enumerate(char_set)}  # char -> index
```

```python
data_dim = len(char_set)
hidden_size = len(char_set)
num_classes = len(char_set)
sequence_length = 10 # 임의로 적당한 값 설정
learning_rate = 0.1
```

```python
dataX = []
dataY = []
```

```python
for i in range(0, len(sentence) - sequence_length):
    x_str = sentence[i:i + sequence_length]
    y_str = sentence[i + 1:i + sequence_length + 1]
    print(i, x_str, '->', y_str)
    
    x = [char_dic[c] for c in x_str] # x str -> index
    y = [char_dic[c] for c in y_str] # y str -> index
    
    dataX.append(x)
    dataY.append(y)

batch_size = len(dataX)
```

```python
# One-hot 인코딩
X_one_hot = tf.one_hot(dataX, num_classes)
Y_one_hot = tf.one_hot(dataY, num_classes)
```

```python
print(X_one_hot.shape)  # (170, 10, 25)
print(Y_one_hot.shape)  # (170, 10, 25)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.
             LSTM(units=hidden_size, input_shape=(sequence_length, X_one_hot.shape[2]), return_sequences=True))
tf.model.add(tf.keras.layers.LSTM(units=num_classes, return_sequences=True))
tf.model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(units=num_classes, activation='softmax')))
tf.model.summary()
tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=learning_rate),
                 metrics=['accuracy'])
```

```python
tf.model.fit(X_one_hot, Y_one_hot, epochs=100)
```

```python
results = tf.model.predict(X_one_hot)
for j, result in enumerate(results):
    index = np.argmax(result, axis=1)
    if j is 0:  # 문장을 만들기 위해 첫 번째 결과는 모두 출력
        print(''.join([char_set[t] for t in index]), end='')
    else:  # 첫 번째 이후 결과들은 마지막 글자만 출력
        print(char_set[index[-1]], end='')
```
