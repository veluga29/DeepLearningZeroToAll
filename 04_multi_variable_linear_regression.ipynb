```python
import numpy as np
import tensorflow as tf
```

```python
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
tf.model = tf.keras.Sequential()
```

```python
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3))  # input_dim=3은 3개의 변수 입력을 의미
tf.model.add(tf.keras.layers.Activation('linear'))  # linear activation이 기본값이기 때문에, 생략해도 되는 코드임
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5))
```

```python
tf.model.summary()
```

```python
history = tf.model.fit(x_data, y_data, epochs=100)
```

```python
y_predict = tf.model.predict(np.array([[72., 93., 90.]]))
```
